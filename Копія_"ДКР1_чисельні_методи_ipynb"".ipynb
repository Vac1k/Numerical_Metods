```python
import pandas as pd
import numpy as np
import math
```

```python
def diff(f, x, delta = 1e-6):
  return (f(x+delta) - f(x-delta))/(2*delta)
```

# Метод дихотомії


```python
e = 0.0001
f = lambda x: 2 ** x + np.e ** (-x) - 5 * x + 1
f_one = lambda x: 2 ** x * np.log(2) - math.e ** (-x) - 5
f_two = lambda x: 2 ** x * np.log(2) ** 2 + math.e ** (-x)



phi = lambda x: x - 0.04 * (x ** 3 - 2 * x + 4)

a_start = 0
b_start = 1
```

```python
# f = lambda x: 0.3 * math.cos(x)**2 - math.log(x) + 2
# f_one = lambda x: -0.6 * np.cos(x) * np.sin(x) - 1 / x

# a_start = 8.5
# b_start = 8.9
```

```python
data_for_print = {'α':[], 'f(α)':[], 'β':[], 'f(β)':[],'x*':[], 'f(x*)':[], '|β-α|':[], '|β-α|<ε':[]}

a = a_start
b = b_start

is_end = False
while not is_end:
  data_for_print['α'].append(round(a, 4))
  data_for_print['f(α)'].append(round(f(a), 4))
  data_for_print['β'].append(round(b, 4))
  data_for_print['f(β)'].append(round(f(b), 4))
  x = (a+b)/2
  data_for_print['x*'].append(round(x, 4))
  data_for_print['f(x*)'].append(round(f(x), 4))
  err = abs(b-a)
  data_for_print['|β-α|'].append(round(err, 4))
  is_end = err < e
  data_for_print['|β-α|<ε'].append(is_end)
  if f(a)*f(x) >= 0:
    a = x
  if f(b)*f(x) >= 0:
    b = x
```

```python
pd.DataFrame(data_for_print)
```

# Метод хорд

```python
data_for_print = {'α':[], 'f(α)':[], 'β':[], 'f(β)':[],'x_k':[], 'f(x_k)':[], '|x_k - x_k-1|':[], '|x_k - x_k-1|<E':[]}
delta = math.inf

a = a_start
b = b_start

def delta_adding(xk_old, immoveble):
  if xk_old == immoveble:
    xk = xk_old
  else:
    xk = xk_old - f(xk_old)*(xk_old-immoveble)/(f(xk_old)-f(immoveble))

  data_for_print['x_k'].append(round(xk, 4))
  data_for_print['f(x_k)'].append(round(f(xk), 4))
  delta = abs(xk - xk_old)
  data_for_print['|x_k - x_k-1|'].append(round(delta, 4))
  data_for_print['|x_k - x_k-1|<E'].append(delta < e)
  return xk, delta



while delta > e:
  data_for_print['α'].append(round(a, 4))
  data_for_print['f(α)'].append(round(f(a), 4))
  data_for_print['β'].append(round(b, 4))
  data_for_print['f(β)'].append(round(f(b), 4))

  if f(a)*f_two(a) >= 0:
    xk_old = b
    immoveble = a

    xk, delta = delta_adding(xk_old, immoveble)

    b = xk

  elif f(b)*f_two(b) >= 0:
    xk_old = a
    immoveble = b

    xk, delta = delta_adding(xk_old, immoveble)

    a = xk
  else:
    print('error!')

```

```python
pd.DataFrame(data_for_print)
```

# Метод дотичних

```python
data_for_print = {'x0':[], 'f(x0)':[], "f'(x0)":[], 'x_k':[], "|f(x_k-1) / f'(x_k-1)|<E":[]}

a = a_start
b = b_start

is_end = False

x0 = [(i, f(i)*f_two(i)) for i in [a,b]]
if x0[0][1] < 0:
  x0 = x0[1][0]
elif x0[1][1] < 0:
  x0 = x0[0][0]
elif x0[1][1] > x0[0][1]:
  x0 = x0[0][0]
else:
  x0 = x0[1][0]
while not is_end:
  data_for_print['x0'].append(round(x0, 4))
  data_for_print['f(x0)'].append(round(f(x0), 4))
  data_for_print["f'(x0)"].append(round(f_one(x0), 4))
  xk = x0 - f(x0)/f_one(x0)
  data_for_print['x_k'].append(round(xk, 4))
  is_end = abs(f(x0)/f_one(x0))<e
  data_for_print["|f(x_k-1) / f'(x_k-1)|<E"].append(is_end)
  x0 = xk
```

```python
pd.DataFrame(data_for_print)
```

```python
phi = lambda x: x + 0.2 * (f(x))
data_for_print = {'k':[], 'xk':[], "φ(xk)":[], "|x_k+1 - x_k|":[], "|x_k+1 - x_k|<ε":[]}

a = a_start
b = b_start

is_end = False

x0 = a
k = 0
while not is_end:
  data_for_print['k'].append(k)
  k+=1
  data_for_print['xk'].append(x0)
  if(abs(diff(phi, x0))>=1):
    print(f"Похідна від ф у точці {x0} приймає значення {diff(phi, x0)}")
    break
  xk = phi(x0)
  data_for_print['φ(xk)'].append(xk)
  is_end = abs(x0-xk)<e
  data_for_print["|x_k+1 - x_k|"].append(abs(x0-xk))
  data_for_print["|x_k+1 - x_k|<ε"].append(is_end)
  x0 = xk
len(data_for_print['φ(xk)'])
```

# Схема Холецького

```python
def true_print(data):
  display(pd.DataFrame(data))

def hypatia_paste(some_data):
  for i in some_data:
    print('{', end='')
    print('}&{'.join(list(map(str, i))).replace('.',','), end='')
    print('}', end='')
    print('\\\\')

def display_with_hypatia(data):
  true_print(data)
  hypatia_paste(data)

def prepare_data(data):
  data = data.replace("\\","").replace(',','.').replace("{","").replace("}","").split('\n')
  data = np.array([np.array(list(map(float, i.split('&')))) for i in data])
  return data
```

```python
def Choletskogo(data):
    sh = data.shape[0]
    L, U = np.zeros((sh, sh)), np.zeros((sh, sh))

    for i in range(sh):
        L[i, i] = U[i, i] = 1

    for i in range(sh):
        L[i:, i] = data[i:, i] - np.matmul(L[i:, :i], U[:i, i])
        U[i, i:] = 1 / L[i, i] * (data[i, i:] - np.matmul(L[i, :i], U[:i, i:]))

    L = np.round(L, 3)
    U = np.round(U, 3)
    determinant = np.prod([j[i] for i,j in enumerate(L)])

    return L, U, determinant
```

```python
data = '''{1.66}&{0.52}&{0.82}&{0.38}\\
{0.78}&{1.69}&{1.56}&{0.49}\\
{0.76}&{1.44}&{1.61}&{0.46}\\
{0.44}&{0.73}&{0.34}&{1.40}\\'''
data = prepare_data(data)
(L, U, determinant) = Choletskogo(data)


print("L:")
display_with_hypatia(L)
print("U:")
display_with_hypatia(U)

print("Результати:")
print("Наш:", round(determinant, 3))
print("Програмно:", round(np.linalg.det(data), 3))
```

# Метод прогонки

```python
coefs = np.array([
    np.array([ 5,  -1,  0,   0]),
    np.array([ 2,  4.6, -1,  0]),
    np.array([ 0,  2,   3.6, -0.8]),
    np.array([ 0,  0,   3,   4.4]),
]
)
di = np.array([
    2,
    3.3,
    2.6,
    7.2
])
```

```python
bi = [-i[j] for j,i in enumerate(coefs)]
bi = np.array(bi)
bi
```

```python
ai = [0]
ai.extend([i[j] for j,i in enumerate(coefs[1:])])
ai = np.array(ai)
ai
```

```python
ci = [i[j+1] for j,i in enumerate(coefs[:-1])]
ci.extend([0])
ci = np.array(ci)
ci
```

```python
alpha = [ci[0]/bi[0]]
gama = [-di[0]/bi[0]]

for i,b,c,a,d in zip(range(0,len(bi)-2), bi[1:-1],ci[1:-1],ai[1:-1], di[1:-1]):
  alpha.append(c/(b-a*alpha[i]))
  gama.append((a*gama[i] - d)/(b-a*alpha[i]))
print("α =",np.round(alpha,3))
print("γ =",np.round(gama,3))
```

```python
n = len(di)
xi = [0 for _ in range(n)]
xi[-1] = (ai[-1]*gama[-1]-di[-1])/(bi[-1]-ai[-1]*alpha[-1])
```

```python
for i in range(n-2, -1, -1):
  xi[i] = alpha[i]*xi[i+1]+gama[i]
for j,i in enumerate(np.round(xi,3)):
  print(f"x_{j+1} = {i}")
```
